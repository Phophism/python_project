```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('../Akkapop/output/dengue_processed_for_clustering.csv')
```

```python
df.head(10)
```

```python
df = df.replace(to_replace = ['9-May', '14-Oct'],
          value = ['5-9','10-14'])
```

```python
df.head(5)
```

```python
plt.scatter(df['temp_avg'],df['rain_ml'])
```

```python
km = KMeans(n_clusters=5)
km
```

```python
y_predicted = km.fit_predict(df[['temp_avg','rain_ml']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head(100)
```

```python
dfc1 = df[df.cluster ==0]
dfc2 = df[df.cluster ==1]
dfc3 = df[df.cluster ==2]
dfc4 = df[df.cluster ==3]
dfc5 = df[df.cluster ==4]


plt.scatter(dfc1.temp_avg,dfc1['rain_ml'],color='green')
plt.scatter(dfc2.temp_avg,dfc2['rain_ml'],color='red')
plt.scatter(dfc3.temp_avg,dfc3['rain_ml'],color='yellow')
plt.scatter(dfc4.temp_avg,dfc4['rain_ml'],color='blue')
plt.scatter(dfc5.temp_avg,dfc5['rain_ml'],color='black')

plt.xlabel('temp_avg')
plt.ylabel('rain_ml')
plt.legend()
```

```python
df.to_csv('file_name.csv')
```
