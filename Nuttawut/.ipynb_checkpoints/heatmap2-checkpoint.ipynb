```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('thailand_lalong.csv')
df1 = pd.read_csv('C:/Users/Mind/Documents/GitHub/python_project/Akkapop/output/dengue_processed.csv')
```

```python
df1.head()
```

```python
df.head()
```

```python
heatmap_df = pd.merge(df1,df,on='Province')
```

```python
heatmap_df.head(50)
```

```python
fig1 = px.density_mapbox(heatmap_df, lat='latitude', lon='longitude', z='patient_count', radius=20,
                        center=dict(lat=13, lon=100), zoom=5,
                        mapbox_style="stamen-terrain")
fig1.show()
```

```python
fig2 = px.density_mapbox(heatmap_df, lat='latitude', lon='longitude', z='rain_ml', radius=10,
                        center=dict(lat=13, lon=100), zoom=5,
                        mapbox_style="stamen-watercolor")
fig2.show()
```
