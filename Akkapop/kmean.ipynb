```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
```

```python
df = pd.read_csv('../Nattawat/dengue_preprocessing.csv')
df
```

```python
df = df.rename(columns={'province': 'Province'})
df
```

```python
coord = pd.read_csv('../Nuttawut/thailand_lalong.csv')
coord
```

```python
df2 = pd.merge(df,coord,on='Province')
df2
```

```python
np_coord = df2[['latitude', 'longitude','quant_rain']].to_numpy()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=5,
    n_init=10,
    max_iter=300,
    random_state=42
)
```

```python
kmeans.fit(np_coord)
```

```python
kmeans.labels_
```

```python
cluster = pd.DataFrame({'k_n':kmeans.labels_})
cluster
```

```python
df3 = df2.join(cluster)
df3
```

```python
sns.scatterplot(data=df3, x="latitude", y="longitude", hue="k_n" )
```
