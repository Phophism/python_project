```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import mpld3
```

```python
df = pd.read_csv('../Nattawat/dengue_preprocessing.csv')
df
```

```python
df = df.rename(columns={'province': 'Province'})
df
```

```python
coord = pd.read_csv('../Nuttawut/thailand_lalong.csv')
coord
```

```python
df2 = pd.merge(df,coord,on='Province')
df2
```

```python
np_coord = df2[['latitude', 'longitude','quant_rain']].to_numpy()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=4,
    n_init=10,
    max_iter=300,
    random_state=42
)
```

```python
kmeans.fit(np_coord)
```

```python
kmeans.labels_
```

```python
centers =  kmeans.cluster_centers_
centers
```

```python
cluster = pd.DataFrame({'k_n':kmeans.labels_})
cluster
```

```python
df3 = df2.join(cluster)
df3
```

```python
# import map
ruh_m = plt.imread('img/map.jpg')
```

```python
# Set edge
maps = pd.DataFrame(data={"lat":[97.313,105.785], "lon":[20.556,5.468]})
maps
```

```python
# Set map edge
BBox = ( maps['lat'].min(), 
         maps['lat'].max(),
         maps['lon'].min(), 
         maps['lon'].max())
BBox
```

```python
fig, ax = plt.subplots(figsize = (8,7))
hmax = sns.scatterplot(
    data=df3, x="longitude", y="latitude",
    hue="k_n" , palette=['red','green','blue','yellow'], 
    size="dead_perc" , sizes=(100, 200) )
# ax.scatter(df3.longitude, df3.latitude, alpha=0.3, c=df3.k_n, s=df3.dead_perc*1000)
hmax.scatter(centers[:, 1], centers[:, 0], s=centers[:,2], alpha=0.1 , c=['red','green','blue','yellow'])
hmax.set_title('Plotting Spatial Data on Riyadh Map')
hmax.set_xlim(BBox[0],BBox[1])
hmax.set_ylim(BBox[2],BBox[3])
tooltip = mpld3.plugins.PointLabelTooltip(scatter, labels=df3.dead_perc)
hmax.plugins.connect(fig, tooltip)
hmax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
```
